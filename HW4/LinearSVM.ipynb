```python
def hard_mining(num,Train_images, feat_extractor_train, D_train, lb_train,D_valid,lb_valid ):
  D_train_pos = [D_train[i] for (i,label) in enumerate(lb_train) if label == 1]
  D_train_neg = [D_train[i] for (i,label) in enumerate(lb_train) if label == -1]
  clf = make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-5, max_iter=1000))
  clf.fit(D_train, lb_train)

  AP = []
  Obj = []

  for i in range(num):
    A = []  # non support vectors
    for (i,x) in enumerate(D_train_neg):
      if (-1 * (clf.named_steps['linearsvc'].coef_.dot(x) + clf.named_steps['linearsvc'].intercept_ ) < 1 ):
        A.append(i)

    #print("Non - Support vectors in NegD : " + str(len(A)))

    ind = np.random.randint(len(Train_images), size=(500))


    B = None
    for m in range(500):
      img = Train_images[ind[m]]
      rects = detect(img, feat_extractor_train, clf)
      rows,cols = np.where(rects[:4] < 0.4)
      rects = rects[rows]
      #print(rects.shape)
      inference_second_stream(feat_extractor_train, img)
      feats = feat_extractor_train.extract_features(rects[:,:-1])
      feats = feats.detach().to('cpu').numpy()
      
      if B is None:
        B = feats
      else:
        B = np.concatenate((B, feats), axis = 0)

      #print(B.shape)
      

    SV_indices = [j for (j,x) in enumerate(D_train_neg) if j not in A]
    D_train_neg = np.take(D_train_neg, SV_indices, axis = 0)

    #print("Hardest Negative Examples : " + str(len(B)))
    
    #print(np.array(D_train_neg).shape, np.array(B).shape)
    D_train_neg = np.concatenate((D_train_neg, np.array(B)),axis =0)
    lb_train_neg = -np.ones(len(D_train_neg))

    D_train = np.concatenate((D_train_pos,D_train_neg), axis =0) 
    lb_train = np.concatenate((np.ones(len(D_train_pos)), lb_train_neg))

    if len(D_train) >= 5000:
      D_train = D_train[:5000]
      lb_train = lb_train[:5000]
    #clf =  LinearSVC(random_state=0, tol=1e-5, max_iter=1000)
    clf.fit(D_train, lb_train)
    
    W = clf.named_steps['linearsvc'].coef_
    b = clf.named_steps['linearsvc'].intercept_

    y_score = clf.decision_function(D_valid)

    Obj.append((np.linalg.norm(W,'fro')**2 )/2)

    AP.append(average_precision_score(lb_valid, y_score))

  return Obj,AP,clf 
```

```python
def get_train_images(num_img): 
    Train_images = []
    dataset_file = "ContactHands/ImageSets/Main/train.txt"
    with open(dataset_file, "r") as f:
        dataset = f.read().splitlines()
    dataset = (dataset if num_img == -1 else dataset[:num_img])

    for ann_name in tqdm.tqdm(dataset):
        # Get annotation and corresponding image
        xml_path = os.path.join(annotations_pool, ann_name + ".xml")
        image_file, boxes = read_content(xml_path)
        image_path = os.path.join(images_pool, image_file)
        # If image in dataset (train or test)
        if image_file.split(".jpg")[0] in dataset:
            # Read image
            img = cv2.imread(image_path)
            #img, boxes = resize_im_boxes(img, boxes, image_size[0], image_size[1])
            Train_images.append(img)
    
    return Train_images
```

```python
Obj, Ap ,cl = hard_mining(10,Train_imgs,feat_extractor_train, D_train,lb_train,D_valid, lb_valid)
```
